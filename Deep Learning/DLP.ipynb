```python
from keras.datasets import mnist
```

```python
(train_images, train_labels),(test_images, test_labels)=mnist.load_data()
```

```python
train_images.shape
```

```python
from keras import models
from keras import layers
```

```python
dnn_clf = models.Sequential()
dnn_clf.add(layers.Dense(512,activation='relu',input_shape=(28*28,)))
dnn_clf.add(layers.Dense(10,activation='softmax'))
```

```python
dnn_clf.compile(optimizer='rmsprop',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
train_images_reformed=train_images.reshape((-1,28*28))
train_images_reformed=train_images_reformed.astype('int64')/255
test_images_reformed=test_images.reshape((-1,28*28))
test_images_reformed=test_images_reformed.astype('int64')/255
```

```python
from keras.utils import to_categorical
```

```python
train_labels_reformed = to_categorical(train_labels)
test_labels_reformed = to_categorical(test_labels)
```

```python
dnn_clf.fit(train_images_reformed,train_labels_reformed,epochs=5)
```

```python
dnn_clf.evaluate(test_images_reformed,test_labels_reformed)
```
