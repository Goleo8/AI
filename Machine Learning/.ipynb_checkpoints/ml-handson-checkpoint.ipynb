```python
import tarfile
import os
import urllib
```

```python
ROOT_URL="https://raw.githubusercontent.com/ageron/handson-ml/master/"
FILE_LOCATION_PATH="datasets/housing"
FILE_NAME="/housing.tgz"
FILE_LOCATION_URL=ROOT_URL+FILE_LOCATION_PATH+FILE_NAME

def data_retrieve(url=FILE_LOCATION_URL,outputPath=FILE_LOCATION_PATH,fileName=FILE_NAME):
    if not os.path.isdir(outputPath):
        os.makedirs(outputPath)
    urllib.request.urlretrieve(url,outputPath+FILE_NAME)
    dataFile = tarfile.open(outputPath+FILE_NAME)
    dataFile.extractall(outputPath)
    dataFile.close()    
```

```python
data_retrieve()
```

```python
housing=df.copy()
housing['median_income_cat']=np.ceil(housing['median_income']/1.5)
housing['median_income_cat'].where(housing['median_income_cat']<5.0,5.0,inplace=True)
```

```python
from sklearn.model_selection import StratifiedShuffleSplit
ssSplit=StratifiedShuffleSplit(n_splits=1,test_size=0.33)
```

```python
for train_index,test_index in ssSplit.split(housing,housing['median_income_cat']):
    train_data_set = housing.iloc[train_index]
    test_data_set = housing.iloc[test_index]
```
