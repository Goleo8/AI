```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_moons

X, y = make_moons(n_samples=500, noise=0.30, random_state=42)
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
plt.scatter(X[:,0],X[:,1],c=y,cmap='autumn',edgecolors='black',linewidths=1.5,s=100)
```

```python
from sklearn.ensemble import VotingClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
```

```python
rf=RandomForestClassifier()
lr = LogisticRegression()
moon_svc=SVC(probability=True)
vot=VotingClassifier(estimators=[('LR',lr),('RF',rf),('SVM',moon_svc)],voting='soft')
```

```python
from sklearn.metrics import accuracy_score
```

```python
for m in (rf, lr, moon_svc, vot):
    m.fit(X_train,y_train)
    y_test_predict = m.predict(X_test)
    print("accuracy score:%s\n"%(accuracy_score(y_test,y_test_predict)))
```
