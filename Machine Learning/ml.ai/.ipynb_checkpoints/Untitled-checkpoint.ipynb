```python
import numpy as np
import pandas as pd
# pip install seaborn 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dtype = {'DayOfWeek': np.uint8, 'DayofMonth': np.uint8, 'Month': np.uint8 , 'Cancelled': np.uint8, 
         'Year': np.uint16, 'FlightNum': np.uint16 , 'Distance': np.uint16, 
         'UniqueCarrier': str, 'CancellationCode': str, 'Origin': str, 'Dest': str,
         'ArrDelay': np.float16, 'DepDelay': np.float16, 'CarrierDelay': np.float16,
         'WeatherDelay': np.float16, 'NASDelay': np.float16, 'SecurityDelay': np.float16,
         'LateAircraftDelay': np.float16, 'DepTime': np.float16}
```

```python
path = '../datasets/2008.csv.bz2'
flights_df = pd.read_csv(path, usecols=dtype.keys(), dtype=dtype)
```

```python
flights_df.head()
```

```python
def parse(df):
    a=df['ArrDelay']

    tempA=a.map(lambda x: np.nan if x<=0 else x)
    tempA=tempA.dropna()
    a_median=np.median(tempA)
    
    d=df['DepDelay']
    
    tempD=d.map(lambda x: np.nan if x<=0 else x)
    tempD=tempD.dropna()
    d_median=np.median(tempD)
    new_df = pd.DataFrame({'ArrDelay':a_median,'DepDelay':d_median},index=df.index)
    print(new_df)
    return new_df
```
