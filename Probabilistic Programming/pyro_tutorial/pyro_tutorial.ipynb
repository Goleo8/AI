```python
import torch
from torch.autograd import Variable
import pyro
import pyro.distributions as dist

import matplotlib.pyplot as plt
%matplotlib inline
```

## simple example   
Using pyro to construct a normal distribution model, which can generate data belong to normal distribution.

```python
mu=Variable(torch.zeros(1))
sigma=Variable(torch.ones(1))
x=dist.Normal(mu,sigma)
samples=[x.sample().data.tolist()[0] for i in range(1000)]
```

__Draw Histgram__

```python
plt.hist(samples)
```

__Draw Density Graph__

```python
from scipy.stats import gaussian_kde
import numpy as np
density = gaussian_kde(samples)
density.covariance_factor = lambda : .25
xs=np.linspace(-5,5,100)
density._compute_covariance()
plt.plot(xs,density(xs))
plt.show()
```

__Compute log probability__

```python
log_p_xs=[x.log_prob(i).data.tolist()[0] for i in samples]
plt.scatter(samples,log_p_xs)
```
